## 1. `from langchain.utilities import SerpAPIWrapper`

### কি কাজ করে?

* **বাংলা:** SerpAPIWrapper ব্যবহার করে তুমি Google Search API এর মাধ্যমে ইন্টারনেট থেকে রিয়েল-টাইম তথ্য আনতে পারো।
* **English:** This wrapper lets you perform live Google searches using SerpAPI inside your LangChain app.

### কখন ব্যবহার করব?

* যখন তোমার অ্যাপ্লিকেশন রিয়েল-টাইম ওয়ার্ল্ড ওয়াইড ওয়েব থেকে ডাটা আনবে (যেমন নিউজ, ওয়েব সার্চ, ইত্যাদি)।
* যখন তোমার LLM কে বাইরের ওয়েব তথ্যের উপর ভিত্তি করে উত্তর দিতে হবে।

### ছোট্ট উদাহরণ:

```python
from langchain.utilities import SerpAPIWrapper
import os

os.environ["SERPAPI_API_KEY"] = "your_serpapi_key_here"

search = SerpAPIWrapper()
result = search.run("Who is the president of Bangladesh?")
print(result)
```

## 2. `from langchain.utilities import PythonREPL`

### কি কাজ করে?

* **বাংলা:** PythonREPL হলো একটি টুল যা তোমার ল্যাঙ্গচেইন এজেন্টকে পাইথন কোড রান করতে দেয়।
* **English:** PythonREPL lets your LangChain agent execute Python code dynamically during conversation or processing.

### কখন ব্যবহার করব?

* যখন তোমার এজেন্টকে **কোড রান করতে হবে**, যেমন গাণিতিক হিসাব, ডাটা প্রসেসিং, বা যেকোনো প্রোগ্রামিং কাজ।
* কোড পরীক্ষার জন্য বা ডাইনামিক ফাংশনালিটি দরকার হলে।

### ছোট্ট উদাহরণ:

```python
from langchain.utilities import PythonREPL

python_repl = PythonREPL()
output = python_repl.run("2 + 2 * 5")
print(output)  # Output: 12
```
