ЁЯФ╣ **ЁЯЗмЁЯЗз When & Why to Use**:
Use this to define **your own tool**, which an agent can call like a function.

ЁЯЫая╕П **Use case**:

* Custom API calls
* Database queries
* Internal functions (e.g., currency converter, weather info)

ЁЯУМ **Example / ржЙржжрж╛рж╣рж░ржг**:

```python
from langchain.tools import tool

@tool
def greet(name: str) -> str:
    """Say hello to someone by name."""
    return f"Hello, {name}!"

# This `greet` function now becomes a LangChain-compatible tool.
```

ЁЯОп **Use with Agent**:

```python
agent = initialize_agent(
    tools=[greet],
    llm=ChatOpenAI(),
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```
