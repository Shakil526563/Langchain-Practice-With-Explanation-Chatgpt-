## ‚úÖ `get_openai_callback` ‡¶ï‡¶ø?

* **‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ:**
  ‡¶è‡¶ü‡¶æ OpenAI API ‡¶è‡¶∞ **token usage** ‡¶è‡¶¨‡¶Ç **cost tracking** ‡¶ï‡¶∞‡¶æ‡¶∞ ‡¶ú‡¶®‡ßç‡¶Ø ‡¶¨‡ßç‡¶Ø‡¶¨‡¶π‡¶æ‡¶∞ ‡¶π‡ßü‡•§ ‡¶§‡ßã‡¶Æ‡¶æ‡¶∞ ‡¶ï‡ßã‡¶°‡ßá ‡¶ï‡ßã‡¶•‡¶æ‡ßü ‡¶ï‡¶§ token ‡¶ñ‡¶∞‡¶ö ‡¶π‡¶ö‡ßç‡¶õ‡ßá ‡¶∏‡ßá‡¶ü‡¶æ ‡¶¶‡ßá‡¶ñ‡¶§‡ßá ‡¶™‡¶æ‡¶∞‡¶¨‡ßá‡•§

## ‚úÖ ‡¶ï‡¶ñ‡¶® ‡¶¨‡ßç‡¶Ø‡¶¨‡¶π‡¶æ‡¶∞ ‡¶ï‡¶∞‡¶¨‡ßá‡¶®?

* ‡¶Ø‡¶ñ‡¶® **token consumption** ‡¶¶‡ßá‡¶ñ‡¶§‡ßá ‡¶ö‡¶æ‡¶ì
* API ‡¶ï‡¶≤‡ßá‡¶∞ **‡¶ï‡ßã‡¶∏‡ßç‡¶ü ‡¶ï‡¶®‡ßç‡¶ü‡ßç‡¶∞‡ßã‡¶≤** ‡¶ï‡¶∞‡¶§‡ßá ‡¶ö‡¶æ‡¶ì
* ‡¶™‡ßç‡¶∞‡ßã‡¶°‡¶æ‡¶ï‡¶∂‡¶® ‡¶¨‡¶æ ‡¶°‡ßá‡¶≠‡ßá‡¶≤‡¶™‡¶Æ‡ßá‡¶®‡ßç‡¶ü‡ßá ‡¶∞‡¶ø‡¶∏‡ßã‡¶∞‡ßç‡¶∏ ‡¶Ö‡¶™‡¶ü‡¶ø‡¶Æ‡¶æ‡¶á‡¶ú ‡¶ï‡¶∞‡¶§‡ßá ‡¶ö‡¶æ‡¶á

---

## ‚úÖ ‡¶õ‡ßã‡¶ü‡ßç‡¶ü Example:

```python
from langchain.llms import OpenAI
from langchain.callbacks import get_openai_callback

llm = OpenAI(model_name="text-davinci-003")

with get_openai_callback() as cb:
    response = llm("Write a poem about spring in Bangla.")
    print(response)
    print(f"\nTokens used: {cb.total_tokens}")
    print(f"Prompt tokens: {cb.prompt_tokens}")
    print(f"Completion tokens: {cb.completion_tokens}")
    print(f"Total cost (USD): ${cb.total_cost}")
```

## 1. `from langchain.schema import Document`

### üìù ‡¶ï‡¶ø?

* **‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ‡ßü:**

### üìå ‡¶ï‡¶ñ‡¶® ‡¶¨‡ßç‡¶Ø‡¶¨‡¶π‡¶æ‡¶∞ ‡¶ï‡¶∞‡¶¨‡ßá‡¶®?

* ‡¶Ø‡¶ñ‡¶® ‡¶§‡ßÅ‡¶Æ‡¶ø ‡¶ï‡ßã‡¶®‡ßã external text, ‡¶Ø‡ßá‡¶Æ‡¶® PDF, ‡¶ì‡ßü‡ßá‡¶¨ ‡¶™‡ßá‡¶ú, ‡¶¨‡¶æ ‡¶Ö‡¶®‡ßç‡¶Ø ‡¶ï‡ßã‡¶®‡ßã ‡¶∏‡ßã‡¶∞‡ßç‡¶∏ ‡¶•‡ßá‡¶ï‡ßá content ‡¶Ü‡¶®‡¶¨‡ßá
* ‡¶Ü‡¶∞ ‡¶∏‡ßá‡¶á content ‡¶ï‡ßá LangChain-‡¶è‡¶∞ vector stores ‡¶¨‡¶æ retrievers ‡¶è ‡¶¶‡¶ø‡¶§‡ßá ‡¶π‡¶¨‡ßá

## 2. `from langchain.schema import messages`

### üìù ‡¶ï‡¶ø?

* **‡¶¨‡¶æ‡¶Ç‡¶≤‡¶æ‡ßü:**
  `messages` ‡¶π‡¶≤‡ßã ‡¶ï‡¶ø‡¶õ‡ßÅ class (‡¶Ø‡ßá‡¶Æ‡¶® `HumanMessage`, `AIMessage`, `SystemMessage`) ‡¶Ø‡¶æ **chat model** ‡¶è‡¶∞ ‡¶Æ‡ßá‡¶∏‡ßá‡¶ú ‡¶∏‡ßç‡¶ü‡ßç‡¶∞‡¶æ‡¶ï‡¶ö‡¶æ‡¶∞ ‡¶°‡¶ø‡¶´‡¶æ‡¶á‡¶® ‡¶ï‡¶∞‡ßá‡•§

* **English:**
  `messages` are classes used to structure chat conversation data by role: user, AI, or system.

### üìå ‡¶ï‡¶ñ‡¶® ‡¶¨‡ßç‡¶Ø‡¶¨‡¶π‡¶æ‡¶∞ ‡¶ï‡¶∞‡¶¨‡ßá‡¶®?

* ‡¶Ø‡¶ñ‡¶® ‡¶§‡ßÅ‡¶Æ‡¶ø OpenAI ‡¶¨‡¶æ ‡¶Ö‡¶®‡ßç‡¶Ø chat-based LLM ‡¶è‡¶∞ ‡¶∏‡¶æ‡¶•‡ßá **chat-style** ‡¶Æ‡ßá‡¶∏‡ßá‡¶ú ‡¶¨‡¶ø‡¶®‡¶ø‡¶Æ‡ßü ‡¶ï‡¶∞‡¶¨‡ßá
* Chat interaction ‡¶è‡¶∞ ‡¶ú‡¶®‡ßç‡¶Ø structured message ‡¶§‡ßà‡¶∞‡¶ø ‡¶ï‡¶∞‡¶æ‡¶∞ ‡¶∏‡¶Æ‡ßü

### ‚ö° ‡¶â‡¶¶‡¶æ‡¶π‡¶∞‡¶£:

```python
from langchain.schema import messages
from langchain.chat_models import ChatOpenAI

chat = ChatOpenAI(model_name="gpt-4")

human_msg = messages.HumanMessage(content="Hello, how are you?")
response = chat([human_msg])

print(response.content)
```
